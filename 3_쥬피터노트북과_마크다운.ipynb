## 왜 쥬피터 노트북인가?

쥬피터 노트북은 프로그램 코드와 실행 결과, 수식, 그래프를 포함하는 문서를 작성하고 공유할 수 있게 해주는 웹 기반의 응용프로그램입니다.

Note: Jupyter는 Julia, Python, R을 조합한 약자입니다.

연구자들은 보통 논문이나 보고서 파일, 자료분석 프로그램 파일, 프로그램 실행 결과 파일을 별도로 관리해 왔습니다.
일반적으로 논문이나 보고서는 데이터와 데이터 분석 프로그램을 거의 포함하지 않습니다.
따라서 논문이나 보고서를 보는 사람은 데이터 분석 과정에 대해 전혀 알지 못하고 분석 결과를 확인할 길이 없습니다.
이런 문제를 해결하기 위해 등장한 개념이 **reproducible research(재현가능한 연구)**라는 개념이며 쥬피터 노트북은 이를 가능하게 해줍니다.

**Reproducible research**는 데이터 수집(data acquisition), 데이터 처리(data processing), 데이터 분석(data analysis)을 투명하게 공개해야 하고 다른 사람들이 접근해서 연구 결과를 재현할 수 있어야 한다는 개념입니다.
쥬피터 노트북은 글 뿐만 아니라 reproducible research가 요구하는 모든 것을 한 곳에 담을 수 있게 해줍니다.

이번 강의 사용되는 모든 자료는 쥬피터 노트북으로 작성하였으며 GitHub에 올려놓았습니다.
프로그래밍 실습을 위해 이 자료를 자신의 Google Drive에 upload하기 바랍니다.

2. Google Drive에 적절한 폴더를 만든 다음 이 폴더에 강의자료 쥬피터 노트북을 upload합니다. 

## 쥬피터 노트북 생성/사용하기

1. Google Drive의 빈 곳에서 오른쪽 마우스 버턴을 눌러서 나타나는 메뉴에서 `더 보기`를 선택하고 그 다음으로 `Google Colaboratory`를 선택하면 새로운 쥬피터 노트북을 생성할 수 있습니다.

2. 쥬피터 노트북은 셀(cell)로 구성되며, 두 종류의 셀을 주로 사용합니다. 프로그래밍을 위한 **코드 셀**과 문서 작성을 **텍스트(text) 셀**이 그것입니다.

3. `markdown`은 텍스트 셀에서 문서를 작성할 때 사용하는 언어입니다.
즉 쥬피터 노트북에서 문서 부분은 `markdown`을 이용하여 작성합니다.
하지만 `markdown`이 수식을 지원하지 않기 때문에 수식은 `LaTeX` 양식으로 입력합니다.

4. 프로그램은 코드 셀에 입력합니다.

5. 코드 셀에 입력한 프로그램을 실행 또는 재실행할 때는 해당 셀을 선택한 다음 `ctrl+enter`를 누르거 셀의 왼쪽에 나타나는 실행 버턴을 누릅니다.

6. 쥬피터 노트북은 하나의 노트북에 있는 모든 코드 셀에 입력된 프로그램을 하나의 프로그램으로 생각합니다.
하지만 이 프로그램을 셀 별로 실행할 수 있으며, 특정 셀을 수정하는 경우 수정한 셀 이후의 코드 셀만 실행하면 됩니다.
이때 `런타임` 메뉴를 적절히 이용하면 편리하게 작업할 수 있습니다.

7. 셀을 삭제, 수정하거나 문서 내에서 이동할 수 있고, 셀의 타입도 변경할 수 있습니다.

## 마크다운

웹의 등장으로 HTML, XML과 같은 ``markup language``가 등장하였지만 이들은 익히고 사용하는데 상당한 시간이 걸리며 그 문서를 읽기 어렵습니다.
이들과 달리 작성하고 읽기 쉬운 ``markup language``로 등장한 것이 **마크다운(markdown)**입니다.

마크다운을 이용하면 메모장과 같은 간단한 편집기로도 질 좋은 문서를 만들 수 있으며, 적절한 프로그램을 사용하면 HTML, LaTeX 등으로 변환할 수도 있습니다.

### 마크다운의 기본적인 문법


1. 제목줄은 `#` 기호로 시작합니다. `#`의 개수가 많아지면 글자 크기가 작아집니다.

2. 한 줄 이상의 빈 줄이 있으면 문단이 바뀝니다.

3. 밑줄 또는 별표로 감싼 부분은 *이탤릭체*가 됩니다.

4. 두 개의 별표로 감싼 부분은 **굵은 글씨체**가 됩니다.

5. 목록을 만들 때는 목록의 항목을 **별표**로 시작하면 됩니다.
목록의 각 항목을 들여쓰기 하면 중첩된 목록을 만들 수 있습니다.

    * 항목 1
    * 항목 2
    
6. 목록 항목에 일련번호를 붙이고 싶으면 목록의 항목을 별표 대신 **정수**로 시작하면 됩니다.
이때 정수를 반드시 순서대로 줄 필요가 없습니다.

    3. 항목 1
    4. 항목 2
    
7. **url link**를 문서에 포함시키고 싶으면 `[링크이름](url)`라고 하면 됩니다. 마크다운에 대한 자세한 설명을 보고 싶으면 [여기](http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html)나 [여기](https://daringfireball.net/projects/markdown/)를 클릭하시오.

8. 문서에 이미지를 포함시키고 싶으면 `![이미지에대한텍스트](경로를포함한이미지파일이름)`와 같이 하면 됩니다.

![파이썬 로고](https://www.python.org/static/community_logos/python-logo-master-v3-TM.png)

**Note:** 인터넷에 있는 이미지를 쥬피터 노특북에 포함하려면 이미지에서 오른쪽 마우스 버턴을 클릭한 다음 ``이미지 주소 복사``를 클릭하고, 이를 쥬피터 노트북에서 붙여넣기 하면 된다.

10. 수식은 LaTeX의 문법에 따라 입력합니다.
본문의 수식은 \$로 감싸고 독립적인 수식은 \$\$로 감쌉니다.
예를 들어 $x \in R$에 대해서 다음 함수를 생각해봅시다.

$$
f(x) = ax^2 + bx + c
$$

LaTeX에 관한 도움말은 [여기](http://www.andy-roberts.net/writing/latex)를 참고하기 바랍니다.

9. 표를 작성할 때 열의 시작과 끝은 파이프(`|`)로 표시합니다.
표의 첫 줄에는 열 이름을 기재하고, 다음 줄에는 각 열의 정렬을 지정합니다.
각 열의 폭은 자동적으로 결정되며, 열의 시작 파이프 다음에 콜론(`:`)을 두면 이 열은 좌측 정렬, 열의 끝을 알리는 파이프 앞에 콜론을 쓰면 해당 열은 우측 정렬됩니다.
만약 열의 시작과 끝 파이프에 모두 콜론을 쓰면 가운데 정렬됩니다.
(정렬은 지원되지 않는 경우가 있습니다.)
