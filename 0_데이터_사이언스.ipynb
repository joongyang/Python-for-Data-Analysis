## 데이터 사이언스의 등장

1962년 John Tukey 교수는 지금의 데이터 사이언스에 해당하는 분야를 **데이터 분석(data analysis)**이라고 불렀습니다.
C. F. Jeff Wu 교수는 1985년 베이징 Chinese Academy of Sciences에서의 강의에서 **통계학의 새로운 이름으로 데이터 사이언스**를 제안하였습니다.

학계에서 데이터 사이언스란 용어가 널리 사용된 것은 2000년 이후입니다.
예를 들어, 2002년에 학술지 *Data Science Journal*이 창간되었으며, 2003년에는 *The Journal of Data Science*가 창간되었습니다.
2014년 미국통계학회는 Statistical Learning and Data Mining 분과를 Statistical Learning and Data Science로 분과 이름을 바꾸었습니다.

일반인에게 **데이터 사이언스(data science)**라는 용어가 알려지게 된 계기는 2012년 Thomas H. Davenport와 D J Patil이 Havard Business Review에 개제한 "Data Scientist: The Sexiest Job of the 21st Century"라는 논문이라고 할 수 있습니다.
이 논문은 많은 신문과 잡지에서 언급되었습니다.



## 데이터 사이언스는 무엇인가?

데이터 사이언스는 주로 규모가 큰 데이터에서 지식을 도출하고, 이 지식을 문제 해결에 응용하는 것을 목적으로 하는 여러 분야가 융합된 학문입니다.
구체적으로 데이터 사이언스는 **해결하고자 하는 문제를 정형화**하고, **데이터를 준비**하고, **데이터를 분석**하고, 이로부터 **해결책을 찾아내고 제시**하는 일을 합니다.
이런 일을 하기 위해서는 컴퓨터 과학, 통계학, 정보 과학, 수학 등에 대한 지식이 요구됩니다.

데이터 사이언스는 새로운 분야가 아니며 그저 통계학의 새로운 이름이라고 하는 사람들도 있으며, 데이터 사이언스가 디지털 자료(digital data)에 중점을 두기 때문에 통계학과는 다르다고 주장하는 사람들도 있습니다.
하지만 데이터의 크기나 컴퓨터의 사용 여부에 의해 데이터 사이언스와 통계학이 다르다고 할 수 없으며 전통적 통계학에서 분기한 통계학의 새로운 응용분야로 보는 것이 주류 의견입니다.

## 데이터 사이언스의 세부 분야

* 빅 데이터(Big Data)

* 기계학습(machine learning, 머신 러닝, 통계적 학습)

* 심층학습(deep learning, 딥 러닝)

* 인공지능(artificial intelligence, AI)
