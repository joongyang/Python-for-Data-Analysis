## 파이썬을 사용하는 여러 가지 방법

### 대화형으로 파이썬 사용하기

가장 기본적인 파이썬 사용법은 **대화형(interactive mode)**입니다.
대화형은 파이썬 프로그램 문장을 입력하면 바로 그 실행 결과를 보여주는 방식입니다.
사용자는 이 결과를 보고 후속 프로그램 문장을 입력할 수 있습니다.
파이썬을 대화형으로 사용하려면 명령 프롬프트에서 `python` 또는 `ipython` 명령을 내리면 됩니다.
두 명령 중에서 `ipython` 명령을 권장합니다.

```python
F:\python> ipython
```

이후에 `ln [1]:` **프롬프트(prompt)**가 나타나면 성공적으로 파이썬이 실행된 것이며 파이썬 프로그램 문장을 실행하고 그 결과를 바로 볼 수 있습니다.

간단한 예로 다음과 같이 화면에 인사말을 출력하도록 `print() 함수`를 실행할 수 있습니다.

```python
ln [1]: print('Hello, Python!!!')
ln [2]: print('안녕 파이썬!!!')
```

대화형 파이썬 사용을 종료하려면 다음과 같이 `exit()` 함수를 실행하면 됩니다.

```python
ln [200]: exit()
```

### 통합개발환경 Spyder에서 파이썬 사용하기

명령 프롬프트에서 작업 디렉토리로 이동한 다음 다음과 같이 `spyder` 명령을 내리면 통합개발환경 Spyder가 실행됩니다.

```python
F:\python> spyder
```

Spyder 통합개발환경은 프로그램 작성과 실행에 필요한 편집기, 파일 탐색기, 파이썬을 대화형으로 사용할 수 있는 환경과 각종 문서와 도움말 등을 제공합니다.
데이터 분석을 제외한 대부분의 프로그래밍 작업은 이 환경에서 수행하는 것이 좋습니다.

### 쥬피터 노트북으로 파이썬 사용하기

파이썬 프로그램과 실행 결과를 포함하는 문서를 작성하고 배포하려면 **쥬피터 노트북**(Jupyter Notebook)을 사용하는 것이 바람직합니다.
특히 데이터 분석용으로 파이썬을 사용할 때 가장 적합한 환경을 제공하는 것이 쥬피터 노트북입니다.
쥬피터 노트북은 웹 환경에서 문서 작성과 프로그래밍을 동시에 할 수 있는 환경을 제공합니다.
뿐만 아니라 쥬피터 노트북을 `LaTeX`, `HTML` 등의 파일로 변환할 수 도 있습니다.

쥬피터 노트북에서 문서를 작성할 때는 **마크다운**(markdown)이라는 간단한 **마크업**(markup) 언어를 사용합니다.
마크다운 사용법은 [Markdown Cheatsheet](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)를 참고하기 바랍니다.

쥬피터 노트북을 실행하려면 명령 프롬프트에서 쥬피터 노트북 파일을 저장할 디렉토리로 이동한 다음 명령을 내립니다.

```python
F:\python> jupyter notebook
```

그러면 웹 브라우저가 열리면서 쥬피터 노트북를 작성할 수 있는 환경이 나타납니다.
이 환경에서 디렉토리와 여러 종류의 파일을 생성하고 변경하고 수정하고 삭제할 수 있습니다.

쥬피터 노트북을 실행한 명령 프롬프트에서 `ctrl+C`를 누르면 쥬피터 노트북이 종료됩니다.

## Google Drive에서 쥬피터 노트북 사용하기

**자신의 컴퓨터에 파이썬을 설치하지 않더라도 인터넷 연결만 있으면 파이썬을 사용할 수 있는 방법이 있습니다.
Google이나 Microsoft와 같은 기업들이 파이썬을 무료로 사용할 수 있게 해놓았기 때문입니다.**
여기서는 Google에서 파이썬을 사용하는 방법에 대해 설명하겠습니다.
**Web browser에서 Google에 로그인하고, Google Drive에 접속하면 Colaboratory app을 사용하여 언제든지 쥬피터 노트북 작업을 할 수 있으며 생성된 쥬피터 노트북은 자동적으로 Google Drive에 저장됩니다.**

다음은 Google Drive에서 쥬피터 노트북을 실행하는 방법입니다.


1. Web browser를 실행합니다.

2. Google에 로그인합니다.

3. Google Drive에 접속합니다.

4. Google Drive에서 쥬피터 노트북을 생성할 디레토리도 이동합니다.

5. 오른쪽 마우스 버턴을 누르면 나타나는 메뉴에서 **`더보기`**를 선택하고, 다음으로 **`연결할 앱 더보기`**를 선택합니다. **Colaboratory를 검색하고 설치**합니다. (이 작업은 한 번만 하면 됩니다.)

6. **기존 쥬피터 노트북을 열 때**: 쥬피터 노트북을 선택한 다음 오른쪽 마우스 버턴을 누르면 나타나는 메뉴에서 **`연결 앱`**을 선택한 다음 **`Google Colaboratory`**를 선택합니다.

7. **새 쥬피터 노트북을 생성할 때**: Google Drive의 빈 곳에서 오른쪽 마우스 버턴을 눌러서 나타나는 메뉴에서 **`더 보기`**를 선택한 다음 **`Google Colaboratory`**를 선택합니다.
