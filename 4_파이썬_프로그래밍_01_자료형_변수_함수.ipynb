# 프로그램은 무엇인가?

**프로그램은 기본적으로 데이터를 처리(data processing)하기 위해 작성**합니다.
**데이터의 값의 모음**이며, **값을 처리한다는 것은 값을 읽고, 쓰고, 저장하고, 연산하는 일을 포괄**합니다.
그런데 처리해야 하는 값은 여러 종류가 있습니다.
**값의 종류에 따라 처리하는 방법이 달라지므로 값의 종류를 이해하는 것은 프로그래밍에서 가장 기본적인 사항**입니다.

## 자료형과 값

파이썬을 비롯한 대부분의 프로그래밍 언어가 기본적으로 이해하고 처리할 수 있는 값은 **정수**(integer, int), **실수**(float), **문자열**(string, str), **논리값**(boolean, bool) 4 종류입니다.
프로그래밍에서는 값의 종류를 **자료형**, **데이터 타입**(data type) 또는 간단히 **타입**(type)이라고 합니다.

우리가 원하는 일을 수행하는 프로그램을 작성하기 위해서는 정수, 실수, 문자열, 논리값 외에도 복소수, 유리수, 날짜, 시각, 시간, 공간 상의 점, 벡터, 행렬, 색깔, 소리, 자동차, 비행기, 배, 게임 캐릭터 ... 같은 많은 자료형이 필요합니다.
이 중 일부는 이미 만들어져 **프로그래밍 언어에 내장**(built-in)되어 있거나 **표준 라이브러리**(standard library)로 배포되고 있습니다.
그렇지 않은 자료형은 프로그래머가 필요에 따라 만들어서 사용해야 합니다.

**프로그래밍 언어에 내장되어 있다는 말은 별다른 조처없이 이런 값을 처리할 수 있다는 말이고, 표준 라이브러리로 배포된다는 말은 언어 자체가 이런 값을 처리할 수 는 없지만 언어와 함께 배포되는 표준 라이브러리를 사용하면 이런 값을 처리할 수 있다는 말입니다.**

**값를 처리하기 위해서는 일단 프로그램에 값을 기재할 수 있어야 합니다.
값을 기재하는 방법은 자료형에 따라 다릅니다.**

아래 프로그램에 기재된 값은 순서대로 정수, 실수, 문자열, 논리값 각각 2개씩입니다.
문자열은 반드시 **따옴표**로 감싸야 합니다.
**작은 따옴표**(single quotation mark)나 **큰 따옴표**(double quotation mark) 어느 것을 사용해도 좋으나 일관성은 유지해야 합니다.
참고로 논리값은 두 가지 밖에 없으며 ``True``와 ``False``가 그것입니다.

일반적으로  **프로그램은 한 개 이상의 문장으로 구성**됩니다.

## 변수

**프로그램을 작성할 때 프로그램 곳곳에 값을 직접 기재하는 것은 바람직하지 않습니다.**
값이 크거나 복잡할 경우 기재할 때 오류를 범하기 쉽고, 다른 값으로 변경해야 할 경우에도 려러 곳에서 수정해야 하는 번거로움이 생기기 때문입니다.

프로그래밍에서도 수학에서와 유사하게 **변수를 만들어서 값을 저장하고 프로그램은 변수를 사용해서 작성하는 것이 좋은 방법**입니다.
**변수에 값을 저장할 때가 아니면 프로그램에 값을 직접 기재하는 것은 반드시 피해야 합니다.**

다음은 **변수를 만들면서 값을 저장하는 문장**입니다.
각 문장은 ``=`` 왼쪽에 기재된 이름을 가진 변수를 만들고 이 변수에 ``=`` 오른쪽의 값을 저장합니다.
여기서 ``=``은 수학과 달리 오른쪽의 값을 왼쪽 변수에 저장하는 일을 하는데 **대입 연산자**(assignment operator)라고 합니다.
이처럼 특정한 일을 수행하는 기호를 **연산자**(operator)라고 합니다.
대입 연산자 외에도 다양한 연산자가 있으며 이들 연산자를 익혀야 프로그램을 작성할 수 있습니다.

**Note: 파이썬 프로그램의 각 문장은 반드시 첫 번째 열부터 입력해야 합니다. 들여쓰기는 정해진 경우에만 사용해야 합니다.
들여쓰기 할 때에도 반드시 4칸 들여쓰기를 해야 합니다.**

```python
first_name = 'Joongyang'
last_name = 'Park'
height = 184.5
weight = 76.5
nb_friends = 5
is_ok = True
```

대입 연산자는 이미 만들어진 변수에 새로운 값을 저장할 때도 사용됩니다.
어떤 값이 저장된 변수에 새로운 값을 저장하면 기존 값을 없어지고 새 값이 저장됩니다.
이런 것을 **덮어쓴다(overwrite)**라고 합니다.

```python
first_name = 'Jy' # 덮어쓴다
weight = 78.3
```

**변수를 만든다는 것은 값을 저장할 공간을 마련한다**는 의미입니다.
여러 개의 값을 사용해서 프로그램을 작성하기 때문에 이들 값을 저장한 공간을 식별하기 위해 변수는 서로 다른 이름을 가져야 합니다.
대입 연산자는 이 공간에 값을 저장하는 일을 하는 연산자라고 할 수 있습니다.

**Note: 값을 저장하는 공간이 바로 메모리(memory)입니다.**

변수를 만들 때 뿐만 아니라 이미 만든 변수를 이용할 때도 변수 이름을 사용합니다.
위에서 `first_name`, `last_name`, `height`, `weight`, `nb_friends` 등이 변수의 이름인데 프로그램에서 변수 이름을 사용하면 해당 변수에 저장된 값이 그 자리에 들어가게 됩니다.



## 함수(function)

**함수는 특정한 일을 하도록 만들어 놓은 독립적인 작은 프로그램**입니다.
함수도 여러 개 있을 수 있기 때문에 변수와 같이 서로 다른 이름을 가지고 있습니다.
함수 이름에 ``()``를 붙이는 것을 프로그램에서는 **함수를 호출(call) 또는 실행(invoke)**한다고 합니다.
여기서 중요한 한 가지 사항은 프로그램을 작성하면 다른 프로그램을 작성할 때 이용할 수 있다는 것입니다.

**함수는 호출될 때만 실행되어 지정된 일을 합니다.**
함수를 호출할 때는 함수가 지정된 일을 할 때 필요한 데이터를 제공하면서 호출해야 합니다.
이런 데이터는 함수 이름 뒤에 오는 ``()`` 속에 기재하는데 이를 **인자(argument)**라고 합니다.
**인자는 함수가 처리할 값을 전달하는 통로**이며 함수는 인자를 통해 전달받은 값을 처리하는 것입니다.
인자의 개수, 인자를 기재하는 방법은 함수를 작성할 때 결정됩니다.
따라서 이미 만들어진 함수를 사용할 때는 함수가 어떤 방식으로 작성되었는지를 설명하는 문서를 참조해야 합니다.

**``print()``** 함수는 이미 작성되어 프로그래밍 언어에 내장되어 있으므로 인자를 기재하는 방법은 고정되어 있습니다.
``print()`` 함수에는 여러 개의 인자를 기재할 수 있으며, 인자가 2개 이상이면 인자 사이에  쉼표(``,``)를 기재해서 인자를 구분해야 합니다.
``print()`` 함수는 인자를 통해 전달받은 값을 지정된 곳(화면이나 파일)에 출력하는 일을 합니다.
기본적으로 화면에 출력합니다.

```python
print(first_name)
print(last_name)
print(weight)
print(first_name, last_name, height, weight, nb_friends)
print(is_ok)
```

변수에 저장된 값의 자료형을 알고 싶을 때는 파이썬 언어에 내장된 **``type()``** 함수를 이용합니다.
변수 이름을 인자로 주고 ``type()`` 함수를 실행하면 그 변수에 저장된 값의 자료형을 알 수 있습니다.



```python
type(last_name)
```

```python
print(nb_friends)
```

```python
type(nb_friends)
```

```python
nb_friends = '진주시'
type(nb_friends)
```

아래 예와 같이 함수 내에 함수가 들어갈 수 있습니다.
이때 내부의 함수가 먼저 실행되며, 내부 함수의 실행 결과가 외부 함수의 인자가 되어 외부 함수가 실행됩니다.
이는 작은 일을 하는 함수를 작성하고 이들를 결합하여 복잡한 일을 하는  프로그램을 작성할 수 있음을 보여줍니다.
이것이 프로그램 특히 함수를 작성하는 기본적인 철학입니다.

```python
print(type(first_name))  # y=f(x), z=g(y), z=g(f(y))
print(type(weight))
```

**대부분의 함수는 지정된 일을 할 뿐만 아니라 일을 한 다음 지정한 결과를 호출한 곳으로 보내줍니다.
(아무 결과를 보내주지 않는 함수도 있습니다.)
함수가 호출한 곳으로 보내주는 결과를 함수의 실행 결과 또는 return value라고 합니다.**
이런 함수의 실행 결과를 변수에 저장하여 후속 데이터 처리에 사용할 수 있습니다.
함수의 실행를 변수에 저장할 때도 대입연산자 ``=``를 사용합니다.

```python
fn_type = type(first_name)
wgt_type = type(weight)
print(fn_type, wgt_type)
```

## 파이썬과 다른 프로그래밍 언어의 차이점

**파이썬은 변수에 자료형을 지정하지 않습니다.
파이썬은 값에 자료형을 지정합니다.**
위의 예처럼 type() 함수에 변수 이름을 인자로 주더라도 이 결과는 변수의 자료형이 아니고 변수에 저장된 값의 자료형입니다.
따라서 한 변수에 정수를 저장했다가 문자열을 저장할 수 있습니다.
대부분의 다른 프로그래밍 언어는 변수에 자료형을 지정합니다.
따라서 변수에 저장할 수 있는 값에 제약이 따릅니다.
즉 변수의 자료형에 해당하는 값만 저장할 수 있습니다.
파이썬에서는 변수의 자료형이 정해져 있지 않고 프로그램이 실행되는 동안 계속 변할 수 있다는 것이 파이썬의 특징 중 하나입니다.
이를 **동적 타이핑(dynamic typing)**이라고 합니다.

## 변수와 함수의 이름

변수의 이름은 해당 **변수에 저장된 값의 의미를 나타내는 이름**을 붙이는 것이 바람직하며, x, y, z, x1, x2, .. 같은 이름은 의미를 전달해주지 않으므로 피해야 한다.
함수의 이름도 해당 함수가 하는 일이 무엇인지를 잘 나타내는 이름을 붙이는 것이 좋습니다.

변수나 함수에 의미있는 이름을 붙여야 한다는 것 외에도 이름을 붙일 때 지켜야 하는 규칙이 있습니다.

1. 이름은 영문자 소문자와 대문자, 밑줄, 숫자로만 지어야 한다. (한글로 이름을 붙이는 것도 가능하지만 그렇게 하지 않는 것이 바람직하다.)
2. 이름의 첫 글자는 숫자일 수 없다.

첫 번째 규칙은 **소문자와 대문자를 서로 다른 문자로 취급한다**는 내용도 포함하고 있습니다.
대소문자를 구별한다는 것은 변수 이름에만 적용되는 것이 아니라 파이썬 프로그램 전반에 적용되는 규칙입니다.
이처럼 대문자와 소문자를 구별하는 것을 ``case-sensitive``하다고 합니다.
그러니까 파이썬은 ``case-sensitive``한 프로그래밍 언어입니다.
