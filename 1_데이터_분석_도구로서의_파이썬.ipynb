## 데이터 사이언스, 빅데이터, 컴퓨터 그리고 인터넷

**데이터 사이언스(Data Science)**는 통계학의 새로운 이름이라고 할 수 있습니다.
통계학에 새로운 이름을 생각하게 된 이유는 **컴퓨터의 발전과 인터넷(Internet)의 등장**이라고 할 수 있습니다.
컴퓨터의 발전과 인터넷(Internet)의 등장으로 인해 사회는 정보화 사회로 진입하였습니다.
정보화 사회가 성숙하면서 대량의 데이터가 축적되었으며, 축적된 데이터를 활용하기 위해 데이터 사이언스가 등장한 것입니다.
데이터를 활용한다는 것을 구체적으로 말하면 데이터를 분석하여 지식을 도출하고 이 지식을 문제 해결에 사용하는 것입니다. 
그리고 데이터를 분석할 때 가장 중요한 도구가 컴퓨터와 이 컴퓨터에서 사용할 데이터 분석 소프트웨어입니다. 

데이터 사이언스는 **빅데이터(Big Data)**, **기계학습(머신러닝, machine learning)**, **심층학습(딥러닝, deep learning)**을 포함하는 분야입니다.
**기계학습**, **심층학습**은 현재 관심을 받고 있는 **인공지능(artificial intelligence, AI)**을 구현하는 가장 중요한 도구이므로 데이터 사이언스 또는 통계학은 인공지능과 밀접한 연관이 있습니다.

지능이 있다면 경험으로부터 지식을 도출할 수 있어야 하고 이 지식을 활용할 수 있어야 합니다.
경험을 기록한 것이 데이터이며, 지식이라는 것은 경험에서 찾아낸 어떤 규칙이나 패턴 또는 법칙 같은 것을 말합니다.
인공지능이란 개념이 등장하기 전부터 통계학은 데이터에 내재한 규칙이나 패턴을 찾아내는 다양한 방법을 개발했기 때문에 현재의 인공지능의 핵심 도구는 통계적 방법일 수 밖에 없습니다.
소위 말하는 데이터 분석이 추구하는 바가 바로 데이터에서 어떤 규칙이나 패턴 또는 법칙을 찾아내는 것이기 때문에 **빅데이터**, **기계학습**, **심층학습**은 결국 데이터 분석이라고 할 수 있습니다.

사회의 정보화가 진행됨에 따라 축적된 데이터 규모가 커지고 있기 때문에 규모가 큰 데이터를 효율적으로 관리하고, 효율적으로 분석하기 위한 여러 가지 방법을 모색하는 것 또한 데이터 사이언스나 빅데이터에서 다루는 문제입니다.
이런 문제는 전통적인 통계학에서는 연구하지 않은 것이므로 데이터 사이언스와 같은 새로운 이름이 등장하게 된 것입니다.



## 패턴, 규칙, 법칙의 예

### 혈액형과 성격

* A형은 순응적이고 자신의 개성을 누른다. 성실하며 인내심이 강하고 모험을 하지 않는다.

* B형은 개성을 강하게 드러내며 행동적이고 정열적이다. 연애에 있어 쉽게 달구어지고 쉽게 깨지기 쉽다. 군인이나 스포츠맨의 리더적인 사람에게 B형이 많다. 자신이 맞는다고 느끼면 무슨 일이 있어도 밀어붙치고 회사에 있어서는 사장 같은 리더타입.

* O형은 변화를 좋아하며 유머가 있다. 로맨티스트이며 여러 분야에서 천재적인 소질을 가지고 있고 비즈니스맨, 외교관으로 출세하는 타입.

* AB형은 몹시 섬세하고 신경질적인 것 같기도 하고 반대로 쉬워보이면서도 차갑다. 두 가지 성격이 섞여있어서 잘 알기 어려운 사람이며 잡기 어려운 성격이라 이성교제에 있어서 제일 애를 먹는다. 너무 여러 가지 많은 것에 신경을 쓰며 낯가림을 많이 한다.

---

* 변수: 성격, 혈액형
* 혈액형이란 변수가 가지는 값: A, B, O, AB
* 성격은 다차원
* 성격을 구성하는 차원(변수): 개성, 리더쉽, 연애에 대한 태도, 변화에 대한 태도, 유머에 대한 태도 등등등

### 성별 연령별 여행

* 여성들은 국내여행지 중 제주도를 가장 선호하고 남성들은 여성과 달리 강원도를 선호한다.

* 여성의 경우 제주도의 예쁜 풍경과 펜션 등의 볼거리에 매력을 느끼고, 남성의 경우 산이나 바다를 통해 즐길 수 있는 레저활동에 관심이 높다.

* 제주도, 강원도, 경기도에 대한 선호도는 거의 전 연령에 걸쳐서 유사한 비중을 차지한다.

* 20대는 부산을 40대는 인천을 여행지로 더 선호한다. 

---

* 성별과 선호하는 여행지
* 연령과 선호하는 여행지
* 남성과 여성은 성별이란 변수가 가지는 값
* 제주도, 강원도, 경기도, 부산, 인천 등은 여행지라는 변수가 가지는 값

### 키, 체중, 소득의 관계

* 키 작은 남성이 장신의 남성보다, 체중이 많이 나가는 여성이 그렇지 않은 여성보다 소득 수준이 낮다는 연구 결과가 나왔다.
영국 엑서터 대학 연구진은 12만명에 가까운 성인 남녀 유전자 정보를 분석한 결과 이같이 결론 내리고 연구 결과를 영국의학저널(BMJ)에 발표했다

* 4 변수: 키, 체중, 성별, 소득
* 남성의 경우: 키와 소득 간에 관계가 있음
* 여성의 경우: 체중과 소득 간에 관계가 있음.
* 즉 성별에 다라 소득과 관계가 있는 신체적 특성이 다르다.

## 데이터 분석을 위한 소프트웨어

**데이터 분석용 소프트웨어에는 상용 소프트웨어와 공개 소프트웨어 두 가지가 있습니다.**
잘 알려진 상용 소프트웨어는 **SAS**, **SPSS**, **Minitab** 등 입니다.
이들 상용 소프트웨어들은 나름의 장단점과 함께 역사를 가지고 있기 때문에 상당수의 사용자를 확보하고 있습니다.
상용 소프트웨어는 최초 구입시 상당한 비용이 들 뿐 아니라, 새로운 버전(version)으로 갱신할 때나 연간 사용 계약을 갱신할 때도 역시 상당한 비용이 듭니다.

**자유 소프트웨어 운동(Free Software Movement)**은 1983년  Richard Stallman이 주도한 것으로 모든 소프트웨어를 누구든지 자유로이 사용할 수 있고, 필요에 따라 변경할 수 있으며, 변경된 소프트웨어를 자유로이 배포할 수 있게 하자는 운동입니다.
1998년 이후 자유 소프트웨어란 용어의 의미를 분명히 하기위해 **오픈 소스 소프트웨어(open source software, OSS, 공개 소프트웨어어)**란 용어를 사용하기 시작하였습니다.
사용자의 수 측면에서도 과거에는 상용 소프트웨어 사용자가 많았지만 2010년 대에 접어들면서 OSS 사용자가 더 많아졌으며 그 차이는 급격하게 벌어지고 있습니다.
현재 가장 많이 사랑받는 데이터 분석용 OSS로는 **R**과 **파이썬(Python)**을 들 수 있습니다.
특히 파이썬은 **객체지향 프로그래밍**을 지원하는 **범용 프로그래밍 언어**이기 때문에 더 많은 관심을 받고 있습니다.

## 파이썬

**파이썬(Python)**은 1991년 2005년부터 2012년까지 구글, 2013년부터 지금까지 Dropbox에서 일하는 네델란드 출신 프로그래머 **귀도 반 로섬(Guido van Rossum, 1956~)**이 발표한 **고급 범용 프로그래밍 언어**입니다.

파이썬은 다음과 같은 특징을 가진 프로그래밍 언어입니다.

1. **플랫폼 독립(platform independent)**이며,
2. **인터프리터(interpreter)** 언어이며,
3. **객체지향 프로그래밍(object-oriented programming)**을 지원하며,
4. **동적 타이핑(dynamically typed)**을 지원한다.

파이썬은 비영리 단체인 **파이썬 소프트웨어 재단(Python Software Foundation , PSF)**이 관리하는 **공개 소프트웨어**입니다.

## 데이터 분석, 기계학습, 심층학습을 위한 파이썬 패키지

**패키지(package)**는 특정 문제를 해결하기 위해 이미 작성된 프로그램의 모음을 일컽는 말입니다.
대부분의 언어는 효율적인 프로그래밍을 위해 다수의 패키지를 제공합니다.
사실 패키지보다 **라이브러리(library)**라는 용어가 더 일반적으로 사용되는 용어입니다.
파이썬 패키지는 디렉토리를 이용한 계층 구조로 여러 개의 **파이썬 파일(모듈, module)**을 체계화한 라이브러리입니다.
각 모듈에는 상수, 함수, 클래스 등이 정의되어 있습니다.
데이터 사이언스를 공부하는 사람들이 필수적으로 익혀야 하는 파이썬 패키지는 다음과 같습니다.

1. [numpy](http://www.numpy.org/): 백터와 행렬 연산을 비롯한 수학적 계산을 위한 패키지
2. [scipy](https://www.scipy.org/): 과학 연구를 위한 패키지
3. [pandas](https://pandas.pydata.org/): 데이터 분석을 위한 자료구조를 지원하는 패키지
4. [statsmodels](http://www.statsmodels.org/stable/index.html): 데이터 분석을 위한 패키지
5. [scikit-learn](http://scikit-learn.org/stable/): 기계학습을 지원하는 패키지
6. [scikit-image](http://scikit-image.org/): 이미지 처리를 위한 패키지
6. [nltk](http://www.nltk.org/), [gensim](https://radimrehurek.com/gensim/): 텍스트 처리를 위한 패키지
7. [konlpy](http://konlpy-ko.readthedocs.io): 한국어 텍스트 처리를 위한 패키지
8. [keras](https://keras.io/), [tensorflow](https://www.tensorflow.org/), [theano](http://deeplearning.net/software/theano/): 심층학습을 위한 패키지
9. [matplotlib](https://matplotlib.org/), [bokeh](https://bokeh.pydata.org/), [seaborn](http://www.statsmodels.org/): 그래프 작성을 위한 패키지

대부분의 파이썬 패키지와 관련 문서는 [The Python Package Index (PyPI)](https://pypi.org/)에 등록되어 있으며 사용자는 PyPI에 등록된 패키지를 자유롭게 사용할 수 있다.
