## 데이터 사이언스, 컴퓨터 그리고 인터넷

전통적인 통계학의 새로운 분야로 데이터 사이언스가 등장하게 된 결정적인 계기는 **컴퓨터의 발전과 인터넷(Internet)의 등장**이라고 할 수 있습니다.
컴퓨터의 발전과 인터넷(Internet)의 등장으로 인해 사회는 **정보화 사회**로 진입하였습니다.
정보화 사회가 성숙하면서 **대량의 데이터가 축적**되었으며, 축적된 데이터를 활용하기 위해 데이터 사이언스가 등장한 것입니다.

데이터를 활용한다는 것을 구체적으로 말하면 **데이터를 분석하여 지식을 도출하고 이 지식을 문제 해결에 사용**하는 것입니다.
그리고 데이터를 분석할 때 가장 중요한 도구가 컴퓨터와 이 컴퓨터에서 사용할 데이터 분석 소프트웨어입니다.

## 데이터 분석을 위한 소프트웨어

**데이터 분석용 소프트웨어에는 상용 소프트웨어와 공개 소프트웨어 두 가지가 있습니다.**
잘 알려진 상용 소프트웨어는 SAS, SPSS, Minitab 등 입니다.
이들 상용 소프트웨어들은 나름의 장단점과 함께 역사를 가지고 있기 때문에 상당수의 사용자를 확보하고 있습니다.
상용 소프트웨어는 최초 구입시 상당한 비용이 들 뿐 아니라, 새로운 버전(version)으로 갱신할 때나 연간 사용 계약을 갱신할 때도 역시 상당한 비용이 듭니다.

**자유 소프트웨어 운동(Free Software Movement)**은 1983년  Richard Stallman이 주도한 것으로 모든 소프트웨어를 누구든지 자유로이 사용할 수 있고, 필요에 따라 변경할 수 있으며, 변경된 소프트웨어를 자유로이 배포할 수 있게 하자는 운동입니다.
1998년 이후 자유 소프트웨어란 용어의 의미를 분명히 하기위해 **오픈 소스 소프트웨어(open source software, OSS, 공개 소프트웨어)**란 용어를 사용하기 시작하였습니다.
사용자의 수 측면에서도 과거에는 상용 소프트웨어 사용자가 많았지만 2010년 대에 접어들면서 OSS 사용자가 더 많아졌으며 그 차이는 급격하게 벌어지고 있습니다.
현재 가장 많이 사랑받는 데이터 분석용 OSS로는 **R**과 **파이썬(Python)**을 들 수 있습니다.
특히 파이썬은 **객체지향 프로그래밍**을 지원하는 **범용 프로그래밍 언어**이기 때문에 더 많은 관심을 받고 있습니다.

## 파이썬

**파이썬(Python)**은 1991년 2005년부터 2012년까지 구글, 2013년부터 지금까지 Dropbox에서 일하는 네델란드 출신으로 프로그래머 **귀도 반 로섬(Guido van Rossum, 1956~)**이 발표한 고급 프로그래밍 언어입니다.
파이썬의 특징은 다음과 같습니다.

1. **플랫폼 독립(platform independent)**이며,
2. **인터프리터(interpreter)** 언어이며,
3. **객체지향 프로그래밍(object-oriented programming)**을 지원하며,
4. **동적 타이핑(dynamically typed)**을 지원한다.

현재 파이썬은 비영리 단체인 **파이썬 소프트웨어 재단(Python Software Foundation , PSF)**이 관리하는 공개 소프트웨어입니다.

## 데이터 분석, 기계학습, 심층학습을 위한 파이썬 패키지

# **패키지(package)**는 특정 문제를 해결하기 위해 이미 작성된 프로그램의 모음을 일컽는 말입니다.
대부분의 언어는 효율적인 프로그래밍을 위해 다수의 **패키지**를 제공합니다.
사실 패키지보다 **라이브러리(library)**라는 용어가 더 일반적으로 사용되는 용어입니다.
파이썬 패키지는 디렉토리를 이용한 계층 구조로 여러 개의 **파이썬 파일(모듈, module)**을 체계화한 라이브러리입니다.
각 모듈에는 상수, 함수, 클래스 등이 정의되어 있습니다.

---

전문가는 전문분야에 관한 패키지를 만들어서 배포하고, 일반 사용자는 이 패키지를 이용하여 자신이 원하는 일을 수행하는 프로그램을 작성합니다.
사용자가 원하는 일을 수행하는 패키지가 없으면 스스로 패키지를 개발해야 합니다.
스스로 패키지를 개발하는 것은 어려운 일이므로 개발되어 배포된 패키지를 최대한 활용하는 것이 매우 현명한 방법입니다.
따라서 자신이 일하는 분야와 관련한 패키지를 잘 알고 있어야 합니다.
데이터 사이언스를 공부하는 사람들이 필수적으로 익혀야 하는 파이썬 패키지는 다음과 같습니다.

1. [numpy](http://www.numpy.org/): 백터와 행렬 연산을 비롯한 수학적 계산을 위한 패키지
2. [scipy](https://www.scipy.org/): 과학 연구를 위한 패키지
3. [pandas](https://pandas.pydata.org/): 데이터 분석을 위한 자료구조를 지원하는 패키지
4. [statsmodels](http://www.statsmodels.org/stable/index.html): 데이터 분석을 위한 패키지
5. [scikit-learn](http://scikit-learn.org/stable/): 기계학습을 지원하는 패키지
6. [scikit-image](http://scikit-image.org/): 이미지 처리를 위한 패키지
6. [nltk](http://www.nltk.org/), [gensim](https://radimrehurek.com/gensim/): 텍스트 처리를 위한 패키지
7. [konlpy](http://konlpy-ko.readthedocs.io): 한국어 텍스트 처리를 위한 패키지
8. [keras](https://keras.io/), [tensorflow](https://www.tensorflow.org/), [theano](http://deeplearning.net/software/theano/): 심층학습을 위한 패키지
9. [matplotlib](https://matplotlib.org/), [bokeh](https://bokeh.pydata.org/), [seaborn](http://www.statsmodels.org/): 그래프 작성을 위한 패키지
