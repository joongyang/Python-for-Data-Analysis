# 비교 연산자와 논리 연산자 그리고 조건문

## 비교 연산자

**비교 연산자(comparison operator)**는 피연산자의 크기를 비교하는 연산자이며, 피연산자는 값 또는 표현식이고 **비교 연산 결과는 논리값**입니다.
비교 연산자에는 `>`, `>=`, `<`, `<=`, `==`, `!=` 가 있습니다.
이들은 순서대로 왼쪽 피연산자의 값이 오른쪽 피연산자의 값보다 큰지(greater than, gt), 같거나 큰지(greater than or equal to, ge), 작은지(less than, lt), 같거나 작은지(less than or equal to, le), 같은지(equal to, eq), 다른지(not equal to, ne)를 알아보고 그 결과를 논리값 `True`, `False`로 주는 연산을 수행합니다.

다음은 반지름이 5인 원의 면적이 80보다 큰지를 알아보는 프로그램입니다.



````python
import math
radius = 5

area = math.pi * math.pow(radius, 2)

is_gt_80 = area <= 80
print(is_gt_80)
````

다음은 둘레가 40보다 크거나 같은지를 알아보는 프로그램입니다.

````python
circumference = 2 * math.pi * radius

is_ge_40 = circumference >= 40
print(is_ge_40)
````

위 예에서 본 `circumference <= 40`나 `(2 * math.pi * radius) <= 40`와 같이 평가 결과가 논리값이 되는 표현식을 **조건**(condition)이라고 합니다.
조건을 평가한 결과가 ``True``이면 **조건을 만족한다**고 하고 ``False``이면 **조건을 만족하지 않는다**고 합니다.

비교 연산을 정수나 실수에만 실행하는 것은 아닙니다.
문자열이나 논리값에 대해서도 비교 연산을 실행할 수 있습니다.
파이썬은 기본적으로 `utf-8` encoding 방식의 unicode 문자열을 사용합니다.
문자열에 대한 비교 연산은 unicode 코드 체계에서 앞에 오느냐 뒤에 오느냐에 따릅니다.
동일한 언어의 경우에는 사전에서 먼저 나오면 작은 값으로 평가합니다. 

````python
# unicode에서 영어가 한글보다 먼저 나온다
"abc" > "가나다"
````

````python
# 영어에서는 대문자가 소문자보다 먼저 나온다
"abc" <= "ABC"
````

````python
"father" > "Father"
````

````python
# 국어 사전에 먼저 나오는 문자가 작은 것으로 간주된다.
"한가" <= "한강"
````

````python
# 논리값에 대해서도 비교 연산이 가능하다.
# True: 1, False: 0
True >= False
````

````python
# 자료형이 다르면 기본적으로 비교 연산이 불가능하다
3.141592 >= "원주율"
````

````python
# 논리값 True는 1, False는 0으로 자동변환되어 비교 연산이 실행된다.
3.141592 >= True
````

````python
True == 1
````

````python
False == 0
````

## 논리 연산자

논리값에 대한 연산을 수행하는 연산자를 **논리 연산자**(logical operator)라고 하는데 `and`, `or`, `not` 3 가지가 있습니다.
논리 연산자 `and`는 두 논리값의 논리곱, `or`는 논리합을 계산해주며, `not`는 논리값을 반대로 만드는 연산자입니다.
논리값이 나오는 표현식 즉 조건이 논리 연산의 피연산자인 경우가 일반적인 논리 연산입니다.

다음은 논리 연산자를 사용한 프로그램입니다.

````python
ans = True and False
print(ans)
````

````python
ans = True or False
print(ans)
````

````python
ans = not False
print(ans)
````

`type()` 함수를 이용하면 논리값의 자료형은 `bool`입을 알 수 있습니다.

````python
print(type(ans))
````

논리 연산자는 두 개 이상의 조건을 결합하여 하나의 조건을 만들 때 자주 사용됩니다.
예를 들어 나이가 20세 이상이고 월급이 200만원 이상이어야 회원 자격이 있다고 할 때 다음은 나이가 35세이고 월급이 195만원인 사람이 회원 자격이 있는지를 알아보는 예입니다.

````python
lb_age = 20                 # 나이의 하한값
lb_salary = 2000000         # 월급의 하한값

age = 35
salary = 1950000

is_eligible = (age >= lb_age) and (salary >= lb_salary)
print(is_eligible)
````

## 조건문

조건문은 조건의 평가 결과가 `True`냐 `False`냐에 따라 다른 계산 또는 일을 하도록 하는 코드를 작성할 때 사용하는 문장입니다.
즉 조건의 만족여부에 따라 다른 일을 수행하도록 프로그램하기 위해 사용하는 문장을 **조건문**(conditional statement)이라 하는데 파이썬의 조건문은 다음과 같이 사용하는 ``if`` 문장입니다.

```python
if 조건:
    조건이 True일 때 실행할 한 줄 이상의 들여쓴 코드
else:
    조건이 False일 때 실행할 한 줄 이상의 들여쓴 코드
```

`if` 문장을 작성할 때는 조건과 `else` 다음에  반드시 `:`(colon)을 써야 하고, 그 다음에 입력하는 한 줄 이상의 코드는 반드시 들여쓰기 해야 합니다.
한 칸 이상을 들여쓰기하면 되지만 반드시 일관성을 유지해야 합니다.
4칸 들여쓰는 것이 관례입니다.
조건이 `False`일 때 수행할 일이 없으면 다음과 같이 `else` 절을 생략할 수 있습니다.

```python
if 조건:
    조건이 True일 때 실행할 한 줄 이상의 들여쓴 코드
```

다음은 반지름이 5인 원의 면적이 80보다 크면 "면적이 80보다 크다."는 메시지를 출력하고 그렇지 않으면 "면적이 80보다 같거나 작다."는 메시지를 출력하는 예입니다.

````python
import math

msg_large = "면적이 80보다 크다."
msg_small = "면적이 80보다 같거나 작다."
lb_area = 80.

radius = 5

area = math.pi * math.pow(radius, 2)

if (area > lb_area):
    msg = msg_large
else:
    msg = msg_small
    
print(msg)
````

다음은 나이가 20세 이상이고 월급이 200만원 이상이어야 회원 자격이 있다고 할 때 다음은 나이가 35세이고 월급이 195만원인 사람이 회원 자격이 있는지를 알아보고 그 결과를 "회원 자격이 있다.", "회원 자격이 없다."는 메시지로 출력하는 예입니다.

````python
msg_yes = "회원 자격이 있다."
msg_no = "회원 자격이 없다."

lb_age = 20
lb_salary = 2000000

age = 35
salary = 1950000

if (age >= lb_age) and (salary >= lb_salary):
    msg = msg_yes
else:
    msg = msg_no
    
print(msg)
````

특정 조건을 만족하지 않을 경우 다른 조건을 만족하는가를 추가로 검사하고자 할 때는 다음과 같이 `else` 대신에 `elif`(else if를 줄인 용어)를 사용할 수 있으며 여러 개의 `elif` 절이 올 수도 있습니다.

```python
if 조건1:
    조건1을 만족할 때 수행할 한 줄 이상의 들여쓴 코드
elif 조건2:
    조건1을 만족하지 않지만 조건2를 만족할 때 수행할 한 줄 이상의 들여쓴 코드
elif 조건3:
    조건1과 조건2를 만족하지 않지만 조건3를 만족할 때 수행할 한 줄 이상의 들여쓴 코드
...
else:
    앞에 기재된 모든 조건을 만족하지 않을 때 수행할 한 줄 이상의 들여쓴 코드
```

나이가 20세 이상이어야 한다는 조건과 월급이 200만원 이상이어야 한다는 조건을 모두 만족하면 정회원 자격이 있고, 두 가지 조건 중에서 한 가지만 만족하면 준회원 자격이 있고, 두 가지 조건을 모두 만족하지 않으면 외원 자격이 없다고 규정되어 있다고 합니다.
다음은 나이가 35세이고 월급이 195만원인 사람이 회원 자격이 있는지를 알아보고 그 결과를 "정회원 자격이 있다.", "준회원 자격이 있다.", "회원 자격이 없다."는 메시지로 출력하는 프로그램입니다.

````python
msg_rmember = "정회원 자격이 있다."
msg_amember = "준회원 자격이 있다."
msg_no = "회원 자격이 없다."

lb_age = 20
lb_salary = 2000000

age = 35
salary = 1950000

if (age >= lb_age) and (salary >= lb_salary):
    msg = msg_rmember
elif (age >= lb_age) or (salary >= lb_salary):
    msg = msg_amember
else:
    msg = msg_no
    
print(msg)
````

`if` 문에서 수행할 일이 또 다른 `if` 문일 수 있다.
이런 `if` 문을 **중첩 `if`문**(nested if)이라고 합니다.
앞의 예를 다음과 같은 중첩 `if` 문으로 프로그램 할 수도 있습니다.

````python
if (age >= lb_age):
    if (salary >= lb_salary):
        msg = msg_rmember
    else:
        msg = msg_amember
else:
    if (salary >= lb_salary):
        msg = msg_amember
    else:
        msg = msg_no
print(msg)
````

````python
if (salary >= lb_salary):
    if (age >= lb_age):
        msg = msg_rmember
    else:
        msg = msg_amember
else:
    if (age >= lb_age):
        msg = msg_amember
    else:
        msg = msg_no
print(msg)
````

$x=5$일 때 함수 $f(x) = \left\{
     \begin{array}{ll}
       1 & \text{if } x \gt 0 \\
       0 & \text{if } x \le 0
     \end{array}
   \right.$를 계산하시오.
   이런 함수를 binary step function이라고 합니다.

````python
x = 5

if (x > 0):
    fx = 1
else:
    fx = 0
    
print(fx)
````

     \begin{array}{ll}
       1 & \text{if } x \gt 0 \\
       -1 & \text{if } x \le 0
     \end{array}
   \right.$를 계산하시오.
   이런 함수를 sign function이라고 한다.

````python
x = -5

if (x > 0):
    fx = 1
else:
    fx = -1
    
print(fx)
````

## `random` 모듈

`random` 모듈에는 난수를 발생하는 여러 함수가 정의되어 있습니다.
자세한 내용은 [`random` 모듈 문서](https://docs.python.org/3/library/random.html)를 참고해야 합니다.

| 함수 | 설명 |
|:------|:-------|
| `seed(int)` | 난수 발생함수의 초기치를 지정하는 함수 |
| `sample(population, k)` | population에서 크기가 k인 표본을 비복원추출하는 함수로 결과는 리스트이다. |
| `choices(population, k=1)` | population에서 크기가 k인 표본을 복원추출하는 함수로 결과는 리스트이다. |
| `randrange(stop)` | `range(stop)`에 속하는 정수 중에서 임의로 하나를 선택해주는 함수 |
| `randrange(start, stop, [step])` | `range(start, stop, [step])`에 속하는 정수 중에서 임의로 하나를 선택해주는 함수 |
| `randint(a, b)` | `randrange(a, b+1)`과 동일함 |
| `random()` | 구간 `[0, 1)`에서 균등분포하는 난수 한 개를 발생하는 함수 |
| `uniform(a, b)` | 구간 `[a, b)`에서 균등분포하는 난수 한 개를 발생하는 함수 |

````python
import random

random.seed(121212)  # 초기치

r = random.uniform(2, 5)   # [2, 5)
print(r)

r = random.uniform(2, 5)   # [2, 5)
print(r)
````

````python
r = random.random()   # random.uniform(0, 1)과 동일함
print(r)
````

````python
random.seed(20200414)

r1 = random.randint(1, 10)  # 1, 2, ..., 10 사이의 정수 중에서 난수 하나를 생성
r2 = random.randint(1, 10)
r3 = random.randint(1, 10)
print(r1, r2, r3)
````

````python
rn = random.random()
print(rn)
rn = random.uniform(5, 10)
print(rn)
````
