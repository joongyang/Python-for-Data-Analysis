# 자료관리와 통계분석

## 자료와 CSV 파일

지금까지 처리할 값이나 자료를 직접 입력하거나 생성하는 예제 프로그램을 작성했습니다.
그러나 데이터를 분석하는 경우 데이터를 직접 입력하는 경우는 거의 없습니다.
대부분의 경우 처리할 데이터가 데이터베이스나 파일에 저장되어 있습니다.
따라서 데이터베이스에 저장된 데이터를 파일로 보낼 수 있으므로 파일에 저장된 데이터를 처리하는 것은 기본적으로 알아야 하는 사항입니다. 

자료를 파일에 저장/공유하는 방법은 여러 가지가 있으나 가장 널리 사용되는 방법은 CSV(comma separated values) 파일로 저장하는 것입니다.
앞에서 소개한 Kaggle과 같은 사이트에서 제공하는 자료도 대부분 CSV 파일입니다.
일반적으로 자료는 2차원 표로 개념화할 수 있는데, CSV 파일은 표의 행(row)을 줄(line)로 대응시킨 텍스트 파일입니다.
그리고 한 행의 여러 값은 쉼표(빈칸 또는 탭)로 구분합니다.

CSV 파일이 텍스트 파일이므로 메모장과 같은 편집기만 있으면 자료를 입력한 CSV 파일을 만들 수 있다.
가장 편리한 방법은 Excel과 같은 spreadsheet 프로그램을 사용하여 자료를 입력/수정한 다음 CSV 형식으로 저장하는 것입니다.

## 데이터 분석과 Pandas

파일에 저장된 데이터를 처리/분석하기 위해서는 파일에 저장된 데이터를 프로그램 내의 자료구조에 담아야 합니다.
일단 자료구조에 담기만 하면 자료구조의 메소드나 통계적 분석법을 구현한 모듈을 이용하여 분석할 수 있습니다.
현재 파이썬을 이용하여 데이터를 분석하는 경우 `Pandas` 패키지의 `DataFrame` 자료구조를 사용하는 것이 표준적인 방법으로 자리잡고 있습니다.
팬더스 패키지는 다음과 같은 좋은 특징을 가지고 있습니다.

1. 자료처리에 용이한 DataFrame 자료구조를 제공합니다.

2. 다양한 형식의 외부 파일을 DataFrame으로 읽어들이는 기능을 제공합니다.

3. DataFrame 자료구조는 기본적인 자료분석과 그래프 작성을 위한 메소드를 가지고 있습니다. 

4. 구현된 많은 기계학습, 심층학습, 빅 데이터 패키지는 데이터가 DataFrame에 저장된 것을 선호합니다.


## CSV 파일 읽기



분석할 자료가 정리되어 CSV 파일에 저장되어 있는 상황에서 강의를 진행합니다.
현실에서는 자료가 데이터베이스에 들어 있거나 Excel 파일과 같은 다른 형식의 파일로 저장되어 있을 수 있습니다.
이때에도 대부분 CSV 파일로 변환할 수 있기 때문에 가장 호환성이 높은 CSV 파일을 기준으로 강의하는 것이 바람직하다고 할 수 있습니다.
특히 Excel 파일을 처리하는 기능도 이번에 배울 Pandas에 구현되어 있으며 CSV 파일을 처리하는 방법과 크게 다르지 않습니다. 

CSV 파일을 읽어 `DataFrame` 객체를 만들 때는 `pandas` 패키지의 `read_csv()` 함수를 사용합니다.
`read_csv()` 함수를 호출할 때 `sep`, `header` **매개변수(parameter)**에 **쉼표가 아닌 값 구분문자**, **열 이름 여부**를 전달할 수 있습니다.

**Note:** 함수를 호출할 때 괄호 속에 기재하는 것을 인자라고 합니다. 인자에 저장된 값은 매개변수에 전달되며 함수는 매개변수를 이용해서 자기가 할 일을 수행합니다. 즉 함수를 작성한 개발자는 매개변수 이름을 결정하고 매개변수를 이용해서 프로그램합니다. 이렇게 개발한 함수를 이용하는 사람은 인자를 통해 매개변수에 값을 전달함으로써 함수를 이용합니다. 함수를 호출할 때 인자만 기재할 수도 있고, 어떤 인자를 어떤 매개변수에 전달하라고 지정할 수도 있습니다. 후자의 방법이 좀 더 명확한 좋은 방법이라고 할 수 있습니다.

패키지를 사용하려면 먼저 패키지를 `import`해야 합니다.
`pandas`는 다음과 같이 `import`하는 것이 관례입니다.
`as` 다음은 패키지에 대한 **별명**(alias)입니다.
패키지에 별명을 붙이지 않아도 되지만 그러면 패키지 이름을 사용해서 프로그램해야 합니다.
패키지 이름이 긴 경우에는 패키지 이름 전체를 기재하는 것 자체가 오류를 일으키는 원인이 될 수 있습니다.

```python
import pandas as pd

meta = pd.read_csv('https://raw.githubusercontent.com/joongyang/python-for-raw-beginners/master/datasets/metabolism.csv')
```

## 객체의 속성과 메소드

**객체(object)**는 객체지향 프로그래밍에서 사용하는 용어입니다.
파이썬은 객체지향 프로그래밍을 지원합니다.
변수가 단순히 값을 저장하는 용도라면 객체는 자료를 저장할 뿐 아니라 이 자료를 처리하는 기능(함수)도 가지고 있습니다.
앞에서 배운 리스트와 같은 자료구조도 객체지향 프로그래밍으로 만들었기 때문에 객체라고 하는 것이 적절한 표현입니다.
따라서 `DataFrame` 객체는 자신이 저장하고 있는 자료와 자료에 대한 정보 그리고 자료를 처리(분석)하는 기본적인 기능을 가지고 있습니다.
이는 자료에 대한 정보를 추출해주거나 자료를 분석하는 함수를 별도로 만들거나 검색하지 않아도 된다는 것을 의미합니다.
따라서 일단 자료가 `DataFrame` 객체로 만들어지면 후속 작업은 상당히 간편해집니다.

자료구조를 설명할 때 언급했지만 다시 한 번 강조합니다.
객체가 하는 일(계산, 연산)은 **객체에 부속된 함수**로 구현되는데 이를 **메소드(method)**라고 합니다.
객체가 보유하고 있는 정보(자료)는 **객체에 부속된 변수**에 저장되는데 이를 **속성(attribute)**이라고 합니다.
따라서 객체로 하여금 어떤 일을 하게 하려면 객체의 메소드를 호출하면 되고, 객체가 저장하고 있는 정보를 보려면 객체의 속성을 조회하면 됩니다.

`DataFrame` 객체가 가지고 있는 속성과 메소드는 다음과 같이 알아볼 수 있습니다.
이름이 밑줄 `_`로 시작하는 속성이나 메소드는 `private member`이므로 이름이 밑줄로 시작하지 않는 `public member`에 집중하면 됩니다.

**NOte:** `private member`와 `public member`도 객체지향 프로그래밍에서 사용하는 용어입니다. `private member`는 개발자가 내부적으로 사용하기 위해 만든 것이고,  `public member`는 개발자가 아닌 사용자를 위한 것입니다. 

```python
for member in dir(meta):
    print(member, end=', ')
        
```

`DataFrame` 객체에 저장된 **자료에 대한 전반적인 정보(meta data)**는 `info()` 메소드로 알아볼 수 있습니다.

```python
meta.info()
```

## 결측치가 아닌 값의 수 조회

각 열에서 **결측치**(missing value)가 아닌 값의 수를 알고 싶으면 `DataFrame` 객체의 `count()` 메소드를 호출하면 됩니다.
결측치는 기본적으로 `None`으로 표시하는데 `count()` 메소드는 각 변수별로 `None`이 아닌 값의 개수를 보여줍니다.

```python
meta.count()
```

특정 열에서 결측치가 아닌 값의 개수는 다음과 같이 알아볼 수 있습니다.

```python
meta.gender.count()
```

```python
meta['gender'].count()
```

`isna()` 메소드는 각 값이 결측치인지 검사한 결과를 논리값으로 줍니다.

```python
meta.isna()
```

```python
meta.weight.isna()
```

다음과 같은 코드로 DataFrame 객체에 새로운 열을 추가하고 결측치를 할당할 수 있습니다.

이때 열 이름을 반드시 `[]`로 감싸서 리스트로 주어야 합니다.

```python
meta = meta.drop(['x'], axis='columns')
meta
```

## 자료의 조회와 수정

`DataFrame` 객체의 `values` 속성을 이용하면 자료 전체를 `numpy`에서 정의한 배열 `ndarray` 자료형으로 접근할 수 있습니다.
`ndarray` 객체에 저장된 값은 `[]` 연산자를 이중으로 사용하여 접근할 수 있습니다.
이때 첫 번째 `[]` 연산자에는 행 번호를 두 번째 `[]` 연산자에는 열 번호를 기재합니다.

대부분의 기계학습, 심층학습 라이브러리들은 `numpy`를 기반으로 하고 있기 때문에  `DataFrame` 객체가 아닌 `ndarray` 객체로 자료를 처리해야 할 때가 있습니다.
이때 `DataFrame` 객체의 `values` 속성을 사용하는 것입니다.

```python
print(meta.values)
print(type(meta.values))
meta.values[0][2]
```

### 연습문제

1. `DataFrame` 객체 `voters`, `wine` 객체의 `values` 속성으로 두 번째 변수의 10번째 값을 조회해보시오.

`DataFrame` 객체의 `values`을 이용하는 것보다 `iloc[]` 메소드를 사용하는 것이 더 편리할 때가 있습니다.
`iloc[]` 메소드를 호출할 때는 접근하고자 하는 행과 열의 번호를 인자로 줍니다.
여러 개의 행과 열을 지정할 때는 **슬라이싱(slicing)**을 사용할 수도 있습니다.

```python
print(meta.iloc[0,2])
print(meta.iloc[0:2, 2])
print(meta.iloc[0:3, 0:2])
print(meta.iloc[:2, 1:])
print(meta.iloc[::3, 2])
```

### 연습문제

1. `DataFrame` 객체 `voters`, `wine`에서 첫 두 변수와 마지막 5행에 해당하는 자료를 추출해보시오.

`DataFrame` 객체의 `iloc[]` 메소드를 사용하여 값을 수정할 수도 있습니다.

```python
meta_cp = meta.copy()     # 복사본을 만들 때 copy() 메소드를 사용한다.
print(meta_cp.iloc[2, 2])
meta_cp.iloc[2, 2] = 30.0
print(meta_cp.iloc[2, 2])
meta_cp.iloc[2, 2] = 36.1
print(meta_cp.iloc[2, 2])
```

## 척도의 지정

컴퓨터로 자료를 처리할 때 입력한 자료로 어떤 계산 또는 연산을 할 수 있는가를 지정하면 예기치 않는 오류를 방지할 수 있습니다.
변수의 **척도(scale)**는 해당 변수를 측정할 때의 정밀도를 나타내는 개념인데 데이터를 분석할 시점에는 어떤 계산이나 연산을 할 수 있는지 나타내는 개념으로 생각하면 됩니다.
변수를 정밀하게 측정하면 할수록 더 많은 게산이나 연산을 할 수 있기 때문입니다.

프로그래밍 언어에서도 이와 유사한 개념으로 **자료형**(data type)이 있습니다.
파이썬에는 수치와 관련한 자료형으로 `int`, `float` 두 가지가 있습니다.
하지만 통계학에서는 수치를 4가지 척도 **명목척도**(nominal scale), **순서척도**(ordinal scale), **구간척도**(interval scale), **비척도**(ratio scale)로 구분하므로 현재 프로그래밍 언어가 이를 다 구현하고 있지 않는 실정입니다.
따라서 올바른 자료분석을 위해서 척도를 지정하는 추가적인 조치를 해야만 합니다.
`pandas` 패키지에 척도를 위한 조치가 구현되어 있지만 구간척도를 지정하는 방법은 없으므로 자료를 분석하는 사람은 구간척도를 갖는 변수에 대해서는 특히 조심해야 합니다.

**Note:** 명목척도를 갖는 변수를 측정한 값으로는 특정 값이 나타난 횟수만 셀 수 있고, 순서척도를 갖는 변수를 측정한 값으로는 특정 값이 나타난 횟수만 세거나 크기 비교를 할 수 있으며, 구간척도를 갖는 변수를 측정한 값으로는 덧셈과 뺄셈을 할 수 있으며,
비척도를 갖는 변수를 측정한 값으로는 사칙 연산을 모두 할 수 있습니다.

대부분의 프로그래밍 언어는 기본적으로 수치를 비척도로 간주하므로 **범주형 변수(categorical variable, 명목척도나 순서척도를 갖는 변수)**는 범주형 변수임을 지정해야 합니다.
이를 위해 `pandas` 패키지는 `Categorical()` 함수를 정의하고 있습니다.
`Categorical()` 함수를 호출할 때 범주형 변수를 측정한 자료를 인자로 전달해야 합니다.
그리고 `Categorical()` 함수의 `categories` 매개변수에는 범주형 변수가 가질 수 있는 값의 목록을 전달하고, `ordered` 매개변수에는 순서척도 여부를 `True` 또는 `False`로 지정해야 합니다.
`ordered` 매개변수에 `True`를 전달하면 순서척도임을 지정하는 것이며 값의 순서는 `categories` 매개변수에 전달한 값의 순서와 같습니다.

한 가지 주의할 점은 `Categorical()` 함수는 원자료의 척도를 변경하지 않고 척도가 변경된 새로운 자료를 만들어 주므로 `Categoricale()` 함수가 반환하는 결과를 저장해야 합니다.

**대사율자료**에서 변수 `gender`의 척도는 명목척도입니다.
따라서 **대사율자료**의 `gender` 열의 척도는 다음과 지정해야 합니다.
