# 쥬피터 노트북(Jupyter Notebook)

쥬피터 노트북은 프로그램 코드와 실행 결과, 수식, 그래프를 포함하는 문서를 작성하고 공유할 수 있게 해주는 웹 기반의 응용프로그램이다.

Note: Jupyter는 Julia, Python, R을 조합한 약자이다.

연구자들은 보통 논문이나 보고서 파일, 자료분석 프로그램 파일, 프로그램 실행 결과 파일을 별도로 관리한다.
일반적으로 논문이나 보고서에는 데이터와 데이터 분석 프로그램이 거의 포함하지 않다.
따라서 논문이나 보고서를 보는 사람은 데이터 분석 과정에 대해 전혀 알지 못하고 분석 결과를 확인할 길이 거의 없다.
이런 문제를 해결하기 위해 등장한 개념이 **reproducible research**라는 개념이며 쥬피터 노트북은 이를 가능하게 해준다.

## 실습 준비: 강의 자료를 download하여 Google Drive에 upload하기

1. [github/joongyang/Python-for-Data-analysis](https://github.com/joongyang/Python-for-Data-Analysis)에서 강의자료롤 작성된 쥬피터 노트북(확장자가 ipynb인 파일)을 download 하고,

2. Google Drive에 적절한 폴더를 만든 다음 이 폴더에 강의자료 쥬피터 노트북을 upload한다. 

## 쥬피터 노트북 사용하기

1. Google Drive의 빈 곳에서 오른쪽 마우스 버턴을 눌러서 나타나는 메뉴에서 `더 보기`를 선택하고 그 다음으로 `Google Colaboratory`를 선택하면 새로운 쥬피터 노트북을 생성할 수 있다.

2. 쥬피터 노트북은 셀(cell)로 구성되며, 두 종류의 셀을 주로 사용한다. 프로그래밍을 위한 코드 셀과 문서 작성을 텍스트(text) 셀이 그것이다.

3. `markdown`은 텍스트 셀에서 문서를 작성할 때 사용하는 언어를 가르키는 용어이다. 그러니까 텍스트 셀에는 `markdown`을 이용하여 문서를 작성한다.
하지만 `markdown`이 수식은 지원하지 않기 때문에  수식은 `LaTeX` 양식으로 입력한다.

4. Python 프로그램은 코드 셀에 입력한다.

5. 코드 셀에 입력한  프로그램을 실행 또는 재실행할 때는  해당 셀을 선택한 다음 `ctrl+enter`를 누르거나 셀의 왼쪽에 나타나는 실행 버턴을 누르면 된다.

6. 쥬피터 노트북은 하나의 노트북에 있는 모든 코드 셀에 입력된 프로그램을 하나의 프로그램으로 생각한다.
하지만 이 프로그램을 셀 별로 실행할 수 있으며, 앞에 위치한  셀이 수정되지 않으면 해당 셀만 실행하면 된다.
하지만 앞에 위치한 셀이 수정되면 수정된 셀 이후의 모든 셀을 다시 실행하여야 한다.
이때 `런타임` 메뉴를 적절히 이용하면 편리하다.

7. 셀을 삭제, 수정하거나 문서 내에서 이동할 수 있고, 셀의 타입도 변경할 수 있다.

## 커널과 R

커널(kernel)은 쥬피터 노트북에 포함된 프로그램 코드를 실행하는 프로그램을 가리키는 용어다.
커널을 변경하면 파이썬이 아닌 다른 언어로 작성된 프로그램을 쥬피터 노트북에서 작성/실행할 수 있다.
현재 사용할 수 있는 커널은 목록은 [여기](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels)에 나와 있다.

## 마크다운

웹의 등장으로 HTML, XML과 같은 ``markup language``가 등장하였지만 이들은 익히고 사용하는데 상당한 시간이 걸리며 그 문서를 읽기 어렵다.
이들과 달리 작성하고 읽기 쉬운 ``markup language``로 등장한 것이 **마크다운**(markdown)이다.

마크다운을 이용하면 메모장과 같은 간단한 편집기로도 질 좋은 문서를 만들 수 있으며, 적절한 프로그램을 사용하면 HTML, LaTeX 등으로 변환할 수도 있다.

## 마크다운으로 문서 작성하는 방법

1. 제목줄: # 기호로 시작한다. #의 개수가 많아지면 글자 크기가 작아진다.

2. 한 줄 이상의 빈 줄이 있으면 문단이 바뀐다.

3. 밑줄 또는 별표로 감싼 부분은 *이탤릭체*가 된다.

4. 두 개의 별표로 감싼 부분은 **굵은 글씨체**가 된다.

5. 목록을 만들 때는 목록의 항목을 **별표**로 시작하면 된다.
목록의 각 항목을 들여쓰기 하면 중첩된 목록을 만들 수 있다.
    * 항목 1
    * 항목 2
    
6. 목록 항목에 일련번호를 붙이고 싶으면 목록의 항목을 별표 대신 **정수**로 시작하면 된다.
이때 정수를 반드시 순서대로 줄 필요가 없다.
    3. 항목 1
    4. 항목 2
    
7. **url link**를 문서에 포함시키고 싶으면 `[링크이름](url)`라고 하면 된다. 마크다운에 대한 자세한 설명을 보고 싶으면 [여기](http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html)나 [여기](https://daringfireball.net/projects/markdown/)를 클릭하시오.

8. 문서에 이미지를 포함시키고 싶으면 `![이미지에대한텍스트](경로를포함한이미지파일이름)`와 같이 하면 된다.

파이썬 로고

10. 수식은 LaTeX의 문법에 따라 입력한다.
본문의 수식은 \$로 감싸고 독립적인 수식은 \$\$로 감싼다.
예를 들어 $x \in R$에 대해서 다음 함수를 생각하자.
$$
f(x) = ax^2 + bx + c
$$
LaTeX에 관한 도움말은 [여기](http://www.andy-roberts.net/writing/latex)를 참고하기 바란다.

9. 표를 작성할 때 열의 시작과 끝은 파이프(`|`)로 표시한다.
표의 첫 줄에는 열 이름을 기재하고, 다음 줄에는 각 열의 정렬을 지정한다.
각 열의 폭은 자동적으로 결정된다.
열의 시작 파이프 다음에 콜론(`:`)을 두면 이 열은 좌측 정렬, 열의 끝을 알리는 파이프 앞에 콜론을 쓰면 해당 열은 우측 정렬한다.
만약 열의 시작과 끝 파이프에 모두 콜론을 쓰면 가운데 정렬된다.
(정렬은 지원되지 않는 경우가 있다.)

| 이름 | 학과 |
|:-----:|:-----|
|박중양|정보통계학과|
| 123 | 678.90|
