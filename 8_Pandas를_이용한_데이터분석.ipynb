# Pandas를 이용한 데이터 분석

## 자료와 CSV 파일 그리고 Excel 파일

지금까지 처리할 값이나 자료를 직접 입력하거나 생성하는 예제 프로그램을 작성했습니다.
그러나 데이터를 분석하는 경우 데이터를 직접 입력하는 경우는 거의 없습니다.
대부분의 경우 **처리할 데이터는 데이터베이스나 파일에 저장되어 있습니다.**
데이터베이스에 저장된 데이터를 파일로 보낼 수 있기 때문에 **파일에 저장된 데이터를 처리하는 것은 데이터 분석가가 반드시 기본적으로 알아야 하는 사항**입니다.

자료를 파일에 저장/공유하는 방식은 여러 가지가 있으나 가장 널리 사용되는 방식은 CSV(comma separated values) 파일로 저장하는 것입니다.
앞에서 소개한 Kaggle과 같은 사이트에서 제공하는 자료도 대부분 CSV 파일입니다.
일반적으로 자료는 2차원 표로 개념화할 수 있는데, CSV 파일은 표의 행(row)을 줄(line)로 대응시킨 텍스트 파일입니다.
그리고 한 행의 여러 값은 쉼표(빈칸 또는 탭)로 구분합니다.

CSV 파일은 텍스트 파일이므로 메모장과 같은 편집기만 있으면 자료를 입력한 CSV 파일을 만들 수 있습니다.
가장 편리한 방법은 Excel이나 Google spreadsheet와 같은 프로그램을 사용하여 자료를 입력/수정한 다음 CSV 형식으로 저장하는 것입니다.

현실적으로 Excel 파일에 자료를 저장하는 경우가 많습니다.
이번 강의에서는 Excel 파일을 중심으로 진행하겠습니다.
하지만 가장 호환성이 좋은 데이터 파일의 형식은 CSV 파일임은 기억하기 바랍니다.


## 외부 파일을 Google Colab에 upload 하기

이번 강의는 구글 Colab에서 쥬피터 노트북을 활용하여 진행하고 있습니다.
구글 Colab에서 쥬피터 노트북을 사용하기 위해 컴퓨터 앞에 앉아 있지만 본인이 사용하는 컴퓨터와 쥬피터 노트북이 실행되는 컴퓨터는 다른 컴퓨터입니다.
따라서 분석하고자 하는 데이터가 저장된 파일을 쥬피터 노트북이 실행되고 있는 컴퓨터에 upload해야 합니다.

외부 파일을 upload하는 방법은 다음과 같습니다.

1. 쥬피터 노트북의 왼쪽 가장자리를 보면 몇 개의 아이콘이 있습니다. 이 중에서 폴더 모양의 아이콘을 클릭합니다. 그러면 아이콘이 펼쳐지면서 쥬피터 노트북이 실행되는 컴퓨터의 저장장치를 탐색할 수 있습니다.

2. 탬색기의 위쪽에 4개의 아이콘이 나타나는데 이 중에서 가장 왼쪽에 있는 아이콘을 클릭하면 upload할 파일을 지정하는 창이 나타납니다.

3. 데이터가 저장된 Excel 파일을 선택하면 경고창이 뜨고 확인 버턴을 무르면 upload가 완료됩니다.

4. upload가 완료되면 upload한 파일 이름이 화면에 나타납니다.

**Note:** 경고창에 뜨는 경고문은 파일을 upload하더라도 쥬피터 노트북 사용이 종료하면 자동적으로 삭제될 것임을 알려주는 것입니다.


위의 방법에 따라 배포한 **metabolism.xlsx, wine.xlsx, X_train.xlsx, Y_train.xlsx, X_test.xlsx, Y_test.xlsx**를 upload하기 바랍니다.

## 데이터 분석과 Pandas

**파일에 저장된 데이터를 처리/분석하기 위해서는 파일에 저장된 데이터를 프로그램 내의 자료구조에 저장해야 합니다.**
데이터를 자료구조에 저장해야 자료구조의 메소드나 통계적 분석법을 구현한 모듈을 이용하여 분석할 수 있습니다.
현재 **파이썬을 이용하여 데이터를 분석하는 경우 `Pandas` 패키지의 `DataFrame` 자료구조를 사용하는 것이 표준적인 방법**으로 자리잡고 있습니다.
Pandas 패키지는 다음과 같은 좋은 특징을 가지고 있습니다.

1. 자료처리에 용이한 DataFrame 자료구조를 제공합니다.

2. 다양한 형식의 외부 파일을 DataFrame으로 읽어들이는 기능을 제공합니다.

3. DataFrame 자료구조는 기본적인 자료분석과 그래프 작성을 위한 메소드를 가지고 있습니다.

4. 구현된 많은 기계학습, 심층학습, 빅 데이터 패키지는 DataFrame을 잘 이해하고 있습니다.

## 객체의 속성과 메소드

**객체(object)**는 객체지향 프로그래밍에서 사용하는 용어입니다.
파이썬은 객체지향 프로그래밍을 지원합니다.
**변수가 단순히 값을 저장하는 용도라면 객체는 자료를 저장할 뿐 아니라 이 자료를 처리하는 기능(함수)도 가지고 있습니다.**
앞에서 배운 리스트와 같은 자료구조도 객체지향 프로그래밍으로 만들었기 때문에 객체라고 하는 것이 적절한 표현입니다.
따라서 **`DataFrame` 객체는 자신이 저장하고 있는 자료와 자료에 대한 정보 그리고 자료를 처리(분석)하는 기본적인 기능을 가지고 있습니다.**
이는 자료에 대한 정보를 추출해주거나 자료를 분석하는 함수를 별도로 만들거나 검색하지 않아도 된다는 것을 의미합니다.
따라서 일단 자료가 `DataFrame` 객체로 만들어지면 후속 작업은 상당히 간편해집니다.

자료구조를 설명할 때 언급했지만 다시 한 번 강조합니다.
객체가 하는 일(계산, 연산)은 **객체에 부속된 함수**로 구현되는데 이를 **메소드(method)**라고 합니다.
객체가 보유하고 있는 정보(자료)는 **객체에 부속된 변수**에 저장되는데 이를 **속성(attribute)**이라고 합니다.
따라서 객체로 하여금 어떤 일을 하게 하려면 객체의 메소드를 호출하면 되고, 객체가 저장하고 있는 정보를 보려면 객체의 속성을 조회하면 됩니다.

`DataFrame` 객체 또는 어떤 객체가 어떤 속성과 메소드를 가지고 있는지 알아보는 것은 어렵지 않습니다.
**객체 이름을 쓰고 마침표를 찍은 다음 잠시 기다리면 속성과 메소드의 이름이 popup 창이 나타납니다.
이 popup 창에서 원하는 속성이나 메소드 이름을 선택하면 해당 속성이나 메소드를 호출하는 코드가 자동적으로 쥬피터 노트북에 나타납니다.
따라서 우리는 각 속성이나 메소드가 어떤 정보 또는 어떤 일을 하는지만 익히면 되는 것입니다.

**Note:** popup 창에서 직육면체 아이콘은 메소드를 나타내고, 스패너 아이콘은 속성을 나타냅니다.

```python
meta.
```

## 열 이름으로 특정 열 추출

열 이름을 사용하면 자료에서 특정 열만 추출할 수 있습니다.
열 이름으로 특정 열을 추출할 때도 객체의 속성을 알아볼 때와 같이 `DataFrame` 객체 이름 다음에 마침표를 찍고 이어서 열 이름을 기재합니다.
열 이름에 빈 칸이 포함되어 있을 때는  `[]` 연산자에 열 이름을 기재하는 방법으로 특정 열만 추출할 수 있습니다.
후자의 방법이 더 일반적인 방법으로 추천합니다.

```python
meta.weight
```

```python
meta[["rate"]]
```

두 개 이상의 열을 추출할 때는 열 이름을 리스트로 만들어 `[]` 연산자에 제공하면 됩니다.

```python
print(meta[["gender", "rate"]])
```

## 결측치가 아닌 값의 수 조회

```python
meta.count()
```

## 척도의 지정

컴퓨터로 자료를 처리할 때 입력한 자료로 어떤 계산 또는 연산을 할 수 있는가를 지정하면 예기치 않는 오류를 방지할 수 있습니다.
변수의 **척도(scale)**는 해당 변수를 측정할 때의 정밀도를 나타내는 개념인데 데이터를 분석할 시점에는 어떤 계산이나 연산을 할 수 있는지 나타내는 개념으로 생각하면 됩니다.
변수를 정밀하게 측정하면 할수록 더 많은 계계산이나 연산을 할 수 있기 때문입니다.

프로그래밍 언어에서도 이와 유사한 개념으로 **자료형**(data type)이 있습니다.
파이썬에는 수치와 관련한 자료형으로 `int`, `float` 두 가지가 있습니다.
하지만 통계학 또는 데이터 분석에서는 수치를 4가지 척도 **명목척도**(nominal scale), **순서척도**(ordinal scale), **구간척도**(interval scale), **비척도**(ratio scale)로 구분하는데 현재의 프로그래밍 언어는 이런 척도를 구분하지 않고 잇습니다.

따라서 올바른 자료분석을 위해서 척도를 지정하는 추가적인 조치를 해야만 합니다.
`pandas` 패키지에 척도를 위한 조치가 구현되어 있지만 구간척도를 지정하는 방법은 없으므로 자료를 분석하는 사람은 구간척도를 갖는 변수에 대해서는 특히 조심해야 합니다.

**Note:** 명목척도를 갖는 변수를 측정한 값으로는 특정 값이 나타난 횟수만 셀 수 있고, 순서척도를 갖는 변수를 측정한 값으로는 특정 값이 나타난 횟수만 세거나 크기 비교를 할 수 있으며, 구간척도를 갖는 변수를 측정한 값으로는 값이 나타난 횟수, 크기 비교 외에 덧셈과 뺄셈을 할 수 있으며,
비척도를 갖는 변수를 측정한 값으로는 추가로 곱셈과 나눗셈을 할 수 있습니다.

**Note:** 변수의 척도는 변수를 측정하는 방법이 결정하며 이를 일부분 나타내는 것이 단위입니다. 하지만 우리가 파일에 저장한 데이터에는 단위가 없습니다. 그러니 척도에 대한 것은 분석하는 사람이 전적으로 처리해야 합니다. 이런 이유로 분석하는 사람은 각 변수의 단위가 무엇인지 파악해야 합니다. 심지어 어떤 방법으로 측정했는지 알아야 할 때도 있습니다.

대부분의 프로그래밍 언어는 기본적으로 수치를 비척도로 간주하므로 **범주형 변수(categorical variable, 명목척도나 순서척도를 갖는 변수)**는 범주형 변수임을 지정해야 합니다.
이를 위해 `pandas` 패키지는 `Categorical()` 함수를 정의하고 있습니다.
`Categorical()` 함수를 호출할 때 범주형 변수를 측정한 자료를 인자로 전달하면 됩니다.
그리고 `Categorical()` 함수의 `categories` 매개변수에는 범주형 변수가 가질 수 있는 값의 목록을 전달하고, `ordered` 매개변수에는 순서척도 여부를 `True` 또는 `False`로 지정합니다.
`ordered` 매개변수에 `True`를 전달하면 순서척도임을 지정하는 것이며 값의 순서는 `categories` 매개변수에 전달한 값의 순서와 같습니다.

한 가지 주의할 점은 `Categorical()` 함수는 원자료의 척도를 변경하지 않고 척도가 변경된 새로운 자료를 만들어 주므로 `Categoricale()` 함수가 반환하는 결과를 저장해야 합니다.

## 표준화와 정규화

기계학습이나 심층학습에서는 원래의 변수를 그대로 사용하지 않고 **표준화(standardization)**하거나 **정규화(normalization)**하여 사용하는 경우가 많습니다.

표준화: (변수의 값 - 변수의 평균)/변수의 표준편차

정규화: (변수의 값 - 변수의 최소값)/(변수의 최대값 - 변수의 최소값)

`sklearn` 패키지의 `preprocessing` 모듈에 정의된 `StandardScaler()` 함수와 `MinMaxScaler()` 함수를 이용하면 표준화와 정규화를 간단히 수행할 수 있습니다.

**Note:** `sklearn` 패키지는 기계학습을 위한 파이썬 패키지이다.

`sklearn` 패키지의 `preprocessing` 모듈에 정의된 `StandardScaler()` 함수와 `MinMaxScaler()` 함수를 호출할 때는 데이터를 `numpy`에서 정의한 배열(array)로 주어야 합니다.
우리가 만든 DataFrame의 `values` 속성이 `numpy` 배열로 저장한 데이터입니다.
그런데 `values` 속성의 데이터는 2차원 배열인데 `StandardScaler()` 함수와 `MinMaxScaler()` 함수는 1차원 배열을 인자로 주어야 합니다.
2차원 배열을 1차원 배열로 만들 때 `numpy` 배열의 `reshape()` 메소드에 -1 과 1을 인자로 줍니다.

**NoteL** 기계학습이나 심층학습을 할 때 가장 자주 일어나는 문제가 바로 자료형 문제입니다. 문서를 잘 읽고 데이터의 자료형을 맞춰 주는 작업을 반드시 수행해야 합니다.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
ss = StandardScaler()
ms = MinMaxScaler()

rate_std = ss.fit_transform(meta.rate.values.reshape(-1, 1))
rate_norm = ms.fit_transform(meta.rate.values.reshape(-1, 1))
```

표준화하거나 정규화한 결과를 DataFrame에 추가할 때는 `numpy` 배열을 DataFrame으로 변환하여 추가해야 합니다.
이때 pandas의 DataFrame() 함수를 이용합니다.

```python
pd.concat([meta, pd.DataFrame(rate_std), pd.DataFrame(rate_norm)], axis=1)
```

## 그래프 작성

`DataFrame` 객체의 `plot()` 메소드로 산점도, 히스토그램, 상자 그림, 꺽은 선 그래프, 원 그래프 등과 같은 기본적인 그래프를 작성할 수 있습니다.
작성한 그래프를 jupyter notebook에 포함시키려면 다음 문장을 한 번 실행해야 합니다.

```python
%matplotlib inline
```

`rate` 열의 값으로 히스토그램을 작성해보기로 하겠습니다.
먼저 `rate` 열의 최대값과 최소값을 구해서 히스토그램을 그릴 범위를 알아보기로 하겠습니다.

```python
print(meta.rate.min())
meta.rate.max()
```

다음은 `rate` 열의 값으로 히스토그램을 작성하기 위해 `plot()` 메소드를 실행하는 예입니다.
`kind` 매개변수에는 작성할 그래프의 종류를 지정합니다.
히스토그램을 작성할 때는 'hist'를 지정하면 됩니다.
`bins` 매개변수에는 계급의 하한값을 리스트로 줍니다.
이때 앞에서 구한 최대값과 최소값이 도움이 됩니다.
`density` 매개변수에는 세로축을 밀도(density)로 할 것인지 도수로 할 것인지를 논리값으로 지정합니다.
`xlim`과 `ylim` 매개변수에는 그래프의 가로축 범위와 세로축 범위를 리스트로 지정합니다.
`color` 매개변수에는 히스토그램의 막대 색깔을 지정합니다.

```python
ax = meta.rate.plot(kind='hist', bins=[900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000],
               density=True,
               xlim=(800, 2000), # ylim=(a, b)
               color='b',
               title='Histogram of Metabolism Rate')
ax.set_xlabel('metabolism rate(cal/hr)')
ax.set_ylabel('density')
```

히스토그램은 계급 구간을 어떻게 하느냐에 따라 매우 다른 그래프가 나올 수 있습니다.
`bins` 매개변수에 계급 구간을 지정하지 않으면 자동적으로 계급이 정해집니다.

```python
ax = meta.rate.plot(kind='hist',
               density=True,
               xlim=(800, 2000), # ylim=(a, b)
               color='g',
               title='Histogram of Metabolism Rate')
ax.set_xlabel('metabolism rate(cal/hr)')
ax.set_ylabel('density')
```

다음은 상자 그림(box plot)을 작성하는 예입니다.
`kind` 매개변수에 'box'를 지정하기만 하면 됩니다.

```python
meta.rate.plot(kind='box')
```

히스토그램과 상자 그림은 구간척도나 비척도를 갖는 변수를 분석할 때 사용하는 그래프(bar chart)입니다.
범주형 변수를 분석할 때는 도수분포표와 막대그래프를 많이 이용합니다.
이처럼 변수의 척도에 따라 분석하는 방법이 달라집니다.
`value_counts()` 메소드를 사용하면 범주형 변수의 도수분포표를 쉽게 작성할 수 있습니다.
그리고 도수분포표를 막대그래프로 시각화할 수 있습니다.

```python
freq = meta.gender.value_counts()
print(freq)
```

```python
relfreq = meta.gender.value_counts(normalize=True)
print(relfreq, type(relfreq))
```

```python
relfreq.plot(kind='bar', color='c')
```

막대그래프를 작성할 때는 `plot()` 메소드의 `kind` 메개변수에 'bar' 또는 'barh'를 지정합니다.

```python
relfreq.plot(kind='barh')
```

`plot()` 메소드로 비척도를 갖는 변수 간의 관계를 시각적으로 알아보기 위한 산점도도 작성할 수 있습니다. (수치적으로 알아보기 위해 계산하는 것이 상관계수입니다.)
이때 `kind` 매개변수에는 'scatter'를 지정하고, `x` 매개변수에는 가로축에 나타낼 열 이름, `y` 매개변수에는 세로축에 나타낼 열 이름을 지정합니다.
